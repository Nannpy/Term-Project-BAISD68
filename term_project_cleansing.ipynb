```python
# Merge 6 year Dataset

import pandas as pd

df1 = pd.read_csv('/content/Dengue_odpc4_2561.csv')
df2 = pd.read_csv('/content/Dengue_odpc4_2562.csv')
df3 = pd.read_csv('/content/Dengue_odpc4_2563.csv')
df4 = pd.read_csv('/content/Dengue_odpc4_2564.csv')
df5 = pd.read_csv('/content/Dengue_odpc4_2565.csv')
df6 = pd.read_csv('/content/Dengue_odpc4_2566.csv')

df_all = pd.concat([df1, df2, df3, df4, df5, df6], ignore_index=True)

df_all['datesick'] = pd.to_datetime(df_all['datesick'], errors='coerce')

df_all = df_all.sort_values('datesick').reset_index(drop=True)

df_all
```

```python
# รวมไฟล์ข้อมูลทั้ง 6 ปี

df_all.to_csv('/content/dengue_2561_2566_merged.csv', index=False)
```

จะแก้ปัญหาโดยใช้ datefine ไปเติมในช่อง datesick เพราะวันวินิจฉัยจะใกล้เคียง กับวันเริ่มป่วยมากที่สุด เเละ datefine ไม่มี missing ปัญหาอาจเกิดจากผู้ป่วยไม่สามารถระบุได้ชัดเจนว่าเริ่มป่วยตั้งเเต่วันไหน



```python
df_all['datedefine'].isnull().sum()
```

```python
# เติม NaN ของ datesick ด้วยข้อมูลของ datefine

df_all['datesick'] = df_all['datesick'].fillna(df_all['datedefine'])
```

```python
df_all['datesick'].isnull().sum()   # หากเป็น 0 เเสดงว่าเติมข้อมูลสำเร็จเเล้ว
```

```python
df_all['datedefine'] = pd.to_datetime(df_all['datedefine'], errors='coerce')
df_all['datefound'] = pd.to_datetime(df_all['datefound'], errors='coerce')
df_all['datedeath'] = pd.to_datetime(df_all['datedeath'], errors='coerce')
```
